Imports

```python
import pandas as pd
import numpy as np

# Transformando os valores em texto para o tipo numérico
from sklearn.preprocessing import LabelEncoder

# Serializando os dados
import pickle

# Decision Tree
from sklearn.tree import DecisionTreeClassifier

# Gráficos
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
with open('/home/uadson/repos/machine-learn/predictive/classification/credit_risk.pkl', 'rb') as data:
    X_credit_risk, y_credit_risk = pickle.load(data) 
```

```python
X_credit_risk
```

```python
y_credit_risk
```

```python
# Treinando o modelo
tree_credit_risk = DecisionTreeClassifier(criterion='entropy')
tree_credit_risk.fit(X_credit_risk, y_credit_risk)
```

```python
# Exibindo os atributos e seus valores de ganho de informação
tree_credit_risk.feature_importances_
```

```python
# Gráficos
predictors = ['História de crédito', 'Dívida', 'Garantias', 'Renda Anual']
image, axis = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))
tree.plot_tree(tree_credit_risk, feature_names=predictors, class_names=tree_credit_risk.classes_, filled=True);
```

```python
# Teste

# História: Boa [0]; Dívida: Alta [0]; Garantias: Nenhuma [1]; Renda Anual > 35 [2];
# História: Ruim [2]; Dívida: Alta [0]; Garantias: Adequada [0]; Renda Anual > 15 [0];

predictions = tree_credit_risk.predict([[0,0,1,2], [2,0,0,0]])
predictions
```
